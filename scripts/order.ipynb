```python
import pandas as pd
```

```python
orders = pd.read_csv('orders_add.csv')
```

```python
orders.head()
```

```python
orders['order Time'] = orders['Order Date'].map(lambda x : x.split(' ')[1])
orders['Order Date'] = orders['Order Date'].map(lambda x : x.split(' ')[0])
orders['Delivered Time'] = orders['Delivered Date'].map(lambda x : x.split(' ')[1])
orders['Delivered Date'] = orders['Delivered Date'].map(lambda x : x.split(' ')[0])
```

```python
orders.head()
```

```python
orders['order Time Hour'] = orders['order Time'].map(lambda x : x.split(':')[0])
orders['Order Date Minute'] = orders['order Time'].map(lambda x : x.split(':')[1])
orders.drop(['order Time'],axis=1,inplace=True)
orders['Delivered Time Hour'] = orders['Delivered Time'].map(lambda x : x.split(':')[0])
orders['Delivered Date Minute'] = orders['Delivered Time'].map(lambda x : x.split(':')[1])
orders.drop(['Delivered Time'],axis=1,inplace=True)
```

```python
orders.head()
```

```python
orders.columns
```

```python
colls = ['order_id','Order Date','Order Date', 'Delivered Date',
       'order Time Hour', 'Order Date Minute', 'Delivered Time Hour',
       'Delivered Date Minute']
```

```python
compile = orders[colls]
```
