```python
order_payment = order.merge(payment, left_on='order_id', right_on='payment_order_id')
```

```python
all_df =order_payment.merge(delivery_driver, left_on='order_id', right_on='delivery_order_id')
```

```python
sample = all_df.sample(frac=0.14 ,random_state=32)
sample.shape
```

```python
sample.head()
```

```python
sample.columns
```

```python
sample.drop(['delivery_order_id_y','payment_order_id_y'],axis=1,inplace=True)
```

```python
sample['delivery_order_id'],sample['payment_order_id'] = sample['delivery_order_id_x'],sample['payment_order_id_x']

sample.drop(['delivery_order_id_x','payment_order_id_x'],axis=1,inplace=True)
```

```python
order_t = sample[or_col]
payment_t = sample[pay_col]
delivery_t = sample[del_col]
```
